## 2. The question: decision, action, cost of a wrong call

## Research Question

Can machine learning rank webpages according to their priority for content refresh using historical SEO and engagement data?

### Decision

Which webpages should the SEO team review first?

### Who takes the action?

SEO specialists or content editors.

### Action

The editor reviews the highest-ranked pages and updates them by improving content, keywords, metadata, and internal links.

### Unit of Analysis

One webpage.

### Output

A ranked list of webpages with a refresh priority score.

### Cost of a Wrong Recommendation

A false positive means a page is refreshed unnecessarily, wasting editor time.

A false negative means an important page is missed, leading to continued traffic decline and lost business opportunities.

### Why Machine Learning?

Simple rules cannot capture the relationships between impressions, clicks, CTR, engagement, rankings, and other SEO signals. Machine learning can combine these signals to produce a better ranked review queue.
