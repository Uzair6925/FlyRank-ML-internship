## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*


This project is a **Ranking (Scoring)** task.

The decision is **which webpages should an SEO specialist review first for a content refresh**. Rather than assigning webpages to categories, the model produces a ranked list so that the pages most likely to need attention appear at the top.

The output supports prioritization, allowing editors to spend their limited time on the webpages with the greatest opportunity for improvement.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*


The starter dataset does not contain an observed "refresh priority" label.

Therefore, this notebook frames the problem as a **ranking task** rather than training a model immediately.

In later stages of the project, the ranking target should come from an **observed future outcome**, such as whether a webpage's performance declined in a later time window after the available features were measured. This follows the principle that the target should be an observed outcome rather than a manually defined rule.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary evaluation metric is **Precision@K**.

Precision@K measures how many of the top K webpages recommended by the model truly deserve review.

A higher Precision@K means the model places the most valuable refresh opportunities near the top of the ranked list, helping SEO specialists focus their efforts effectively.

## 5. Why ML beats a fixed rule here
A simple rule such as "refresh every page older than one year" or "refresh pages with low CTR" ignores many other signals that influence content performance.

SEO performance depends on multiple interacting factors, including search demand, impressions, click-through rate, average position, engagement, content age, and content characteristics.

Machine learning can learn complex relationships among these signals and produce a better ranked review queue than a single hand-written rule.

The output is intended to support human decision-making rather than automate content updates.
